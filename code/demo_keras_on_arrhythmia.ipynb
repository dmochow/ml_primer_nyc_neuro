```python
import numpy
from tensorflow import keras
from keras import layers, regularizers
import scipy.io as sio
```

```python
mat_contents=sio.loadmat('midterm_challenge_data_set.mat')
mat_contents
#mat_contents=sio.loadmat('my_cancer_dataset.mat')
```

```python
xTrain=mat_contents['xTrain']
yTrain=mat_contents['yTrain']
xTest=mat_contents['xTest']
yTest=mat_contents['yTest']
print("xTrain shape = ", xTrain.shape)
print("xTest shape = ", xTest.shape)
print("yTrain shape = ", yTrain.shape)
print("yTest shape = ", yTest.shape)
print(np.unique(yTrain))
```

```python
model = keras.Sequential(
    [
        layers.Input(shape=274),
        layers.Dense(5,activation='relu',kernel_regularizer=regularizers.L1L2(0.01, 0.01)),
        layers.Dense(5,activation='relu',kernel_regularizer=regularizers.L1L2(0.01, 0.01)),
        layers.Dense(1,activation='sigmoid',kernel_regularizer=regularizers.L1L2(0.01, 0.01))
    ]
)

#model = keras.Sequential(
#    [
#        layers.Dense(10,activation='relu',kernel_regularizer=regularizers.L1L2(0.01, 0.01)),
#        layers.Dense(1,activation='sigmoid',kernel_regularizer=regularizers.L1L2(0.01, 0.01))
#    ]
#)
model.summary()
```

```python
model.compile(loss=keras.losses.BinaryCrossentropy(
    from_logits=True), optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(xTrain,yTrain,batch_size=32, epochs=100, validation_split=0)
```

```python
trainScores=model.evaluate(xTrain,yTrain,verbose=0)
print("Train loss:", trainScores[0])
print("Train accuracy:", trainScores[1])
```

```python
testScores=model.evaluate(xTest,yTest,verbose=0)
print("Test loss:", testScores[0])
print("Test accuracy:", testScores[1])
```
