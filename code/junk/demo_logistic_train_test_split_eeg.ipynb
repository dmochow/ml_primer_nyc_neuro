```python
import scipy.io as sio
import numpy as np
```

```python
dataFilename='P300_py.mat'
data=sio.loadmat(dataFilename)
```

```python
data
```

```python
Xtrain = data['Xtrain']
ytrain = data['ytrain']
Xtest = data['Xtest']
ytest = data['ytest']
```

```python
Xtrain.shape
```

```python
Xtrain_2D = Xtrain.reshape(Xtrain.shape[0],-1)
Xtest_2D = Xtest.reshape(Xtest.shape[0],-1)
```

```python
Xtrain_2D.shape
```

```python
ytrain_1D = ytrain[:,1]
ytest_1D = ytest[:,1]
```

```python
ytrain_1D
```

```python
np.mean(ytrain_1D)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=2000,C=0.001).fit(Xtrain_2D,ytrain_1D)
```

```python
model.coef_
```

```python
ytrain_hat = model.predict(Xtrain_2D)
np.mean(ytrain_hat == ytrain_1D)
```

```python
ytest_hat = model.predict(Xtest_2D)
np.mean(ytest_hat == ytest_1D)
```

87% -- that seems good -- but is it?

```python
from sklearn import metrics
py_train_hat = model.predict_proba(Xtrain_2D)[:,1]
py_test_hat = model.predict_proba(Xtest_2D)[:,1]

#calculate AUC of model
auc_train = metrics.roc_auc_score(ytrain_1D,py_train_hat)
auc_test = metrics.roc_auc_score(ytest_1D, py_test_hat)
```
